```python
import pandas as pd
import numpy as np
import scipy.stats as stats
# Read the CSV file into a DataFrame
filepath_2014='/Users/aman/Downloads/RPC11_Input_For_Participants/datasets/constituency_wise_results_2014.csv'
filepath_2019='/Users/aman/Downloads/RPC11_Input_For_Participants/datasets/constituency_wise_results_2019.csv'
filepath_state_code='/Users/aman/Downloads/RPC11_Input_For_Participants/datasets/dim_states_codes.csv'
df_2014=pd.read_csv(filepath_2014)
df_2019=pd.read_csv(filepath_2019)
df_code=pd.read_csv(filepath_state_code)

df_2014
```

### ****Capitalise The Entire Data Set if its a String****

```python
df_2014 = df_2014.map(lambda x: x.capitalize() if isinstance(x, str) and x!="NOTA" else x)
df_2019 = df_2019.map(lambda x: x.capitalize() if isinstance(x, str) and x!="NOTA" else x)
df_2014[['category','party']]=df_2014[['category','party']].map(lambda x:x.upper() if x!='General' and not pd.isna(x) else x)
df_2019[['category','party']]=df_2019[['category','party']].map(lambda x:x.upper() if x!='General' and not pd.isna(x) else x)
df_2019
```

### ****If Candidate Name is NOTA/None of the above. Then Sex,Age,Category column is NA,0 & NA respectively.****

```python
# Define conditions to check for 'none of the above', 'nota' or None in the 'Name' column
conditions = df_2014['candidate'].isin(['None of the above', 'NOTA']) | df_2014['candidate'].isna()

# Update the corresponding rows
df_2014.loc[conditions, ['sex', 'category', 'age']] = ['NOTA', 'NOTA', 0]

df_2014['age'] = df_2014['age'].fillna(0).astype(int)

conditions = df_2019['candidate'].isin(['None of the above', 'NOTA']) | df_2019['candidate'].isna()
df_2019.loc[conditions, ['sex', 'category', 'age']] = ['NOTA', 'NOTA', 0]
df_2019['age'] = df_2019['age'].fillna(0).astype(int)


df_2014
```

### ****Remove Suffix From Category Column to Constituency Name****

```python
# Function to remove suffix from pc_name if it matches category
def remove_suffix(row):
    pc_name = row['pc_name']
    category = row['category']
    
    # Check if pc_name ends with (category) and remove it
    if pc_name.endswith(f" ({category})"):
        return pc_name[:-len(f" ({category})")].strip()
    else:
        return pc_name

# Apply the function to each row
df_2014['pc_name'] = df_2014.apply(remove_suffix, axis=1)
df_2019['pc_name'] = df_2019.apply(remove_suffix, axis=1)
df_2014
```

### ****Mapping Against 2019 Dataset to Cater The Bifurcation of AP in 2014. i.e Alloating Constituency Against Newly Formed TL****

```python
# Create a mapping dictionary from df_2019
mapping_dict = dict(zip(df_2019['pc_name'], df_2019['state']))

# Map the state column in df_2014 using the mapping_dict
df_2014['state'] = df_2014['pc_name'].map(mapping_dict)

df_2014
```

## 5. To find the top 5 candidates based on the margin difference with runners-up in 2014 and 2019,

##### ***5a. Identify the Winner and Runner-Up: Determine the winning and runner-up candidates for each constituency in both 2014 and 2019 based on the total votes received.<br><br>5b. Calculate the Margin Difference: Compute the margin difference between the winner and the runner-up for each election.<br><br>5c. Rank the candidates based on their margin difference and select the top 5.***

```python
# Function to calculate margin differences for a given year
def calculate_margins(df):
    # Sort candidates by votes within each constituency
    sorted_df = df.sort_values(by=['state', 'pc_name', 'total_votes'], ascending=[True, True, False])
    # Group by constituency and extract the top two candidates (winner and runner-up)
    top_two = sorted_df.groupby(['state', 'pc_name']).head(2)
    # Pivot to get winner and runner-up side by side
    pivot = top_two.pivot_table(index=['state', 'pc_name'], values=['candidate', 'total_votes'], aggfunc=list)
    # Calculate margin differences
    pivot['margin_diff'] = pivot['total_votes'].apply(lambda x: x[0] - x[1] if len(x) > 1 else 0)
    return pivot

# Calculate margins for 2014 and 2019
margins_2014 = calculate_margins(df_2014)
margins_2019 = calculate_margins(df_2019)

# # Combine results for both years
# combined_margins = pd.merge(
#     margins_2014[['candidate', 'margin_diff']], 
#     margins_2019[['candidate', 'margin_diff']], 
#     left_index=True, 
#     right_index=True, 
#     suffixes=('_2014', '_2019')
# )

# Find top 5 candidates based on margin difference in 2019 & 2014
top_5_candidates_2014 = margins_2014.sort_values(by='margin_diff', ascending=False).head(5)
print('Top 5 Candidates B/O Margin Difference 2014 :',top_5_candidates_2014)

top_5_candidates_2019 = margins_2019.sort_values(by='margin_diff', ascending=False).head(5)
print('Top 5 Candidates B/O Margin Difference 2019 :',top_5_candidates_2019)


```

## 12. Correlation between postal votes percentage and voter turnout percentage 
## ***Using Pearson's correlation coefficient***

```python
# Calculate postal votes percentage
df_2019['postal_votes_percentage'] = (df_2019['postal_votes'] / df_2019['total_electors']) * 100

# Calculate voter turnout percentage
df_2019['voter_turnout_percentage'] = (df_2019['total_votes'] / df_2019['total_electors']) * 100

# Compute the correlation between postal votes percentage and voter turnout percentage
correlation = df_2019['postal_votes_percentage'].corr(df_2019['voter_turnout_percentage'])

# Determine the strength of the correlation
if correlation > 0.7 or correlation < -0.7:
    correlation_strength = "Strong"
elif correlation > -0.3 and correlation < 0.3:
    correlation_strength = "Weak"
else:
    correlation_strength = "Moderate"

correlation_strength
```
